```python
# 실습문제1
with open("./data.txt","r",encoding="utf-8") as f:
  sum = 0
  lines = f.readlines()
  for line in lines:
    val = int(line.rstrip())
    sum += val
    if val % 2 == 0:
      print("{}:짝수".format(val))
    else:
      print("{}:홀수".format(val))
  print("정수의 갯수:{}".format(len(lines)))
  print("정수의 총합:{}".format(sum))
  print("정수의 평균:{:.2f}".format(sum/len(lines)))
```

```python
# 실습문제2
with open("./text.txt", "r",encoding="utf-8") as f:
  lines = f.readlines()
  str_sum = ""
  for line in lines:
    line = line.replace("\n", " ")
    str_sum += line
  print("총 문장의 수:", len(lines))
  print("총 단어의 수:", len(str_sum.split(" ")))
```

```python
# 실습문제3
import string
with open("./smart_contact.txt","r",encoding="utf-8") as f:
  result = {}
  for line in f:
    line = line.rstrip()
    translator = str.maketrans("", "", string.punctuation)
    line = line.translate(translator)
    words = line.split()
    for word in words:
      if word in result: result[word] += 1
      else: result[word] = 1
  sort_dict = sorted(result.items(), key=lambda item: item[1], reverse=True)
  result = {}
  for _dict in sort_dict:
    result[_dict[0]] = _dict[1]
  print(result)
```

```python
# 실습문제4
with open("./students.txt", "r", encoding="utf-8") as f:
  lines = f.readlines()
  result = {}
  for line in lines:
    line.replace("\n", "")
    name, score = line.split()
    result[name] = int(score)
  sort_dict = sorted(result.items(), key = lambda item: item[0])
  print(result.keys())
  print(result.values())
  print("----- Name : Score -----")
  for _tuple in sort_dict:
    print("{} : {}".format(_tuple[0], _tuple[1]))
```

```python
# 실습문제5
import csv
with open("./dollar.csv", "r", encoding="utf-8-sig") as f:
  csv_data = csv.reader(f)
  won2dollar = []
  for idx, row in enumerate(csv_data):
    for element in row:
      if idx == 0:
        print(element, end=" ")
      elif idx == 1:
        won2dollar.append(float(element))
        print("{:10.1f}".format(float(element)), end=" ")
      elif idx == 2:
        if element == "0":
          print(" " * 4, "변화율", sep="", end=" ")
        else:
          print("{:10.2f}".format(float(element)), end=" ")
    if not idx == 3: print()
  print("==>최저 환율 : {:.1f} 최고 환율 : {:.1f}".format(min(won2dollar), max(won2dollar)))
```

```python
# 실습문제6
def int_input(prompot):
  try:
    value = input(prompot)
    if value.find(".") > 0:
      print("정수가 아닙니다. 다시 입력하세요.")
      return int_input(prompot)
    return int(value)
  except ValueError:
    print("정수가 아닙니다. 다시 입력하세요.")
    return int_input(prompot)

if int_input("숫자를 입력하세요:"): 
  print("정수 입력이 성공했습니다!")
```
